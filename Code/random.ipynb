```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
path = '/Users/zhuyuezx/Desktop/DMD_captures'
```

```python
def filter_bright_range(img: np.ndarray, threshold: int = 100) -> np.ndarray:
    bright_pixels = img > threshold
    # get smallest and largest indices of bright pixels, both in x and y direction
    x_min, x_max = np.min(np.where(bright_pixels)[0]), np.max(np.where(bright_pixels)[0])
    y_min, y_max = np.min(np.where(bright_pixels)[1]), np.max(np.where(bright_pixels)[1])
    return img[x_min:x_max, y_min:y_max]
```

```python
target = 'strip_white_wo_lens'

img = cv.imread(path + '/' + target + '.png', cv.IMREAD_GRAYSCALE)

# crop the centre part of the image
img = img[550: 1100, 500: 1050]

plt.imshow(img, cmap='gray')

cv.imwrite(path + '/' + target + '_cropped.png', img)
```

```python
img = cv.imread(path + '/' + target, cv.IMREAD_GRAYSCALE)
plt.imshow(img, cmap='gray')
```

```python
# get the gradient of the image
sobelx = cv.Sobel(img, cv.CV_64F, 1, 0, ksize=5)
sobely = cv.Sobel(img, cv.CV_64F, 0, 1, ksize=5)
gradient = np.sqrt(sobelx ** 2 + sobely ** 2)
plt.imshow(gradient, cmap='gray')
```
